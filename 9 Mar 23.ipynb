```python
def findz(eqn, result):
    # eqn: [4, 2, 1, 5],
    # result: [16 / 15, - 8/3]
    total = 0
    for i in range(len(result)):
        v = result[i]
        total += v*eqn[i]
    return (eqn[-1] - total)/eqn[-2]
```
