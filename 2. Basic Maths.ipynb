````python
while True:
    print(f"Left: {left}")
    print(f"Right: {right}")
    mid = (left+right)/2
    print(f"Mid: {mid}")
    result = 10**mid
    print(f"Result: {result}")
    diff = abs(result - num)
    print(f"Diff: {diff}")
    if diff <= 0.0001:
        print(f"Success!! log10{num}: {mid}", )
        break
    else:
        if result > num:
            right = mid
        else:
            left = mid
````

````python
def log10(num):
    left = 0
    right = 10
    while True:
        mid = (left+right)/2
        result = 10**mid
        diff = abs(result - num)
        if diff <= 0.0001:
            return mid
        else:
            if result > num:
                right = mid
            else:
                left = mid
````
