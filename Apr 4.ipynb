```python
def final_compound_interest(years, rate_per_year, principle_amount):
    print(f"{years, rate_per_year, principle_amount}")
    if years == 0:
        return principle_amount
    else:
        new_amount = principle_amount + rate_per_year*principle_amount/100 # p *( 1 + r/100)
        left_years = years - 1
        return final_compound_interest(left_years, rate_per_year, new_amount)
```

```python
def find_first_empty_space(board):
    nrows = len(board)
    ncols = len(board[0])
    for i in range(nrows):
        for j in range(ncols):
            if board[i][j] == 0:
                return i, j
    return None, None
```

```python
# Back Tracking

def solve_sudoku(board):
    row, col = find_first_empty_space(board)
    
    if row == None:
        print("Congratulation!")
        return True
        
    for i in range(1, 10): # valid Moves and start trying them one at a time
        if can_place(i, row, col, board): # if the move can be made
            board[row][col] = i # make the move
            if solve_sudoku(board): # repeat
               return True
            else:
               board[row][col] = 0 #back track
    return False
```

```python
# factorial - fact(5) = 5 * 4 * 3 * 2 * 1, fact(n) = n * fact(n-1)
def fact(n):
    print(f"Fact: {n}")
    if n == 0:
        return 1
    return n * fact(n-1)
```

```python
# n * m = m + m + ....n times = m + m * (n-1)
# n is integer
# m is decimal

def mult_int(n, m): 
    if n == 0:
        return 0
    return m + mult_int(n-1, m)
        
```
