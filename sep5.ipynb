```python
# Amitosh
def findMinima3(start,step,itr,stepGrowth):
    x=start
    while(itr>0):
        current = p(x)
        left = p(x - step)
        right = p(x + step)
        if (right > current) and (left > current):
            step = step / 2  
        elif (right > current) and (left < current):
            step = step * stepGrowth  
        elif (left > current) and (right < current):
            step = step * stepGrowth
        if right>current:
            x=x-step
        else:
            x=x+step
        itr=itr-1
    val=p(x)
    print(f"Minimum p(x) is: {val}")
    print(f"Minimum x is: {x}")
    print(f"Step size now is: {step}")
```

```python
def findMinima(p, lr=0.01, epochs=100):
    x = .5
    d = 0.00000001
    for i in range(epochs):
        px = p(x)
        print(x, px)
        dp = (p(x+d) - px) / d
        x = x - lr * dp
    return x
```

```python
def minimizer(f, lr=0.01, epochs=100):
    x1,x2 = 0.5, 0.5
    d = 0.000000001
    for i in range(epochs):
        fx = f(x1, x2)
        print(x1, x2, fx)
        dfx1 = (f(x1+d, x2) - fx)/d
        dfx2 = (f(x1, x2+d) - fx)/d
        print("grads: ", dfx1, dfx2)
        x1 -= lr*dfx1
        x2 -= lr*dfx2
    return x1, x2
```

```python
minimizer(p, lr=.01, epochs=200)
```
