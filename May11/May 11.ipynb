```python
import re
from pprint import pprint

# Regex to capture timestamp, level, thread/context and message
LOG_PATTERN = re.compile(
    r'^(?P<timestamp>\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2},\d{3}) - '
    r'(?P<level>\w+)\s+'
    r'\[(?P<context>.+?)\] - '
    r'(?P<message>.+)$'
)

def parse_zookeeper_log(file_path):
    entries = []
    with open(file_path, 'r') as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            m = LOG_PATTERN.match(line)
            if m:
                entries.append(m.groupdict())
            else:
                # optionally handle lines that don't match
                entries.append({
                    "raw": line
                })
    return entries

if __name__ == "__main__":
    logs = parse_zookeeper_log("zookeeper.txt")
    pprint(logs)
```
