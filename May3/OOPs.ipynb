```python

class TwoPointLine:
    '''This is the Two Point Class by SG'''
    def __init__(self, name="SG"):
        self.name = name
        print(f"The class has been constructed {name}")
    def classMethod(x):
        print(f"This is a class method: {x}")
        
    def fit(self, x, y):
        if len(x) != 2 or len(y) != 2:
            print("Error! Invalid Data")
            return
        x1, x2 = x
        y1, y2 = y
        m = (y2 - y1) / (x2 - x1)
        c = y2 - m*x2
        self.params = (m, c)
    def predict(self, x):
        m, c = self.params
        result = []
        for xx in x:
            yy = m * xx + c
            result.append(yy)
        return result

```
