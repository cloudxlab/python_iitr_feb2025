````python
a = [4,2,3]
b = [3, 2, 4]

def calc_abs_error(a, b):
    abs_error = 0
    for a1, b1 in zip(a, b):
        diff = a1 - b1
        if diff < 0:
            diff = -diff
        abs_error += diff
    
    return abs_error
````

````python
calc_abs_error(a, b)
````

````python
# Find min
arr = [1,2,3,4, 3, -2, 56]

#Approach 1
def mymin(arr):
    mymin = arr[0]
    for e in arr[1:]:
        if e < mymin:
            mymin = e
    return mymin
````

````python
mymin(arr) # Order(n)
````

````python
#Approach2
arr.sort() # Order(n log n)
arr[0]
````

````python
def myhashcode(name):
    myhash = 0
    for c in name:
        myhash = (myhash*35 + ord(c))%1024
    return myhash
````
